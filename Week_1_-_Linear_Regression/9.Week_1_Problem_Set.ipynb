The linear regression algorithm takes into account the data points from a set of data and repeatedly tries to create a line which best fits the dataset provided. It does this by checking how far off the line is from each data point and finalises with the lowest distance between the line and each data point. 

__5. Modify the code block below to adjust $w$ and $b$ to manually fit a Linear Regression line with good intuition or fit.__

````python
x_train = np.array([[-4], [-8], [-10], [-12], [-14]])
y_train = np.array([1, 3, 5, 7, 9])

w = -0.9
b = -3.9

tmp_f_wb = compute_model_output(x_train, w, b,)

plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.title("Title")
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.legend()
plt.show()
````
